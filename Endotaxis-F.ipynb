## Tower of Hanoi graph

```python
def MakeHanoiGraph(n):
    '''
    Constructs a graph for the Tower of Hanoi game with n disks
    Returns:
    Sl = a list of states given as (x,y) coordinates on the graph. state 0 is at (0,0)
    Es = a sparse connectivity matrix among those states, given as a list of pairs [i,j]
    for all the connected states, where i and j are indices into s.
    '''
    def width(n): # side length of the triangle of level n graph
        w=0
        for i in range(n):
            w=2*w+1
        return w
    
    def doHanoi(x,y,n): # make a level n graph with bottom left at (x,y)
        w=width(n) # find the side length of the triangle
        d1=(w-1)//2
        d2=(w+1)//2
        p=(x+d1,y+d1) # add six states near the middle of each side of the triangle
        q=(x+d2,y+d2)
        St.add(p);St.add(q)
        Ed.add((p,q));Ed.add((q,p)) # and their edges both ways
        p=(x+2*w-d1,y+d1)
        q=(x+2*w-d2,y+d2)
        St.add(p);St.add(q)
        Ed.add((p,q));Ed.add((q,p)) 
        p=(x+2*d1,y)
        q=(x+2*d2,y)
        St.add(p);St.add(q)
        Ed.add((p,q));Ed.add((q,p)) 
        if n>1: # call the routine to make the 3 sub-triangles
            doHanoi(x,y,n-1)
            doHanoi(x+d2,y+d2,n-1)
            doHanoi(x+2*d2,y,n-1)
    
    Ed=set() # dictionary of edges
    St=set() # dictionary of states
    
    doHanoi(0,0,n) # recursion
    
    Sl=np.array(list(St)) # list of state coordinates (x,y)
    Sl=Sl[np.argsort(Sl[:,1])] # sort by y
    Sl=Sl[np.argsort(Sl[:,0])] # sort by x
    Lu={tuple(p):i for i,p in enumerate(Sl)} # a dictionary to look up state number by coordinates
    El=list(Ed) # list of edges with coordinates
    Es=np.array([[Lu[p],Lu[q]] for (p,q) in El])
    Es=Es[np.argsort(Es[:,1])] # sort by second state
    Es=Es[np.argsort(Es[:,0])] # sort by first state
    
    return Sl,Es
```

```python
def PlotHanoiGraph(n,ax):
    '''
    Draws the node numbers of Hanoi graph with n levels into axes ax
    '''

    def width(n): # side length of the triangle of level n graph
        w=0
        for i in range(n):
            w=2*w+1
        return w

    Sl,_=MakeHanoiGraph(n)
    ax.set_xlim([0,width(n)*2]);
    ax.set_ylim([0,width(n)]);
    for i,(x,y) in enumerate(Sl):
        ax.text(x,y,str(i));
    ax.axis('off');
    
```

```python
fig,ax=plt.subplots(1,1,figsize=(5,5))
PlotHanoiGraph(4,ax)
```

## Adjacency matrix

```python
def AdjRing(n):
    '''
    Returns adjacency matrix for ring graph with n nodes
    '''
    A=np.zeros((n,n))
    for i in range(n):
        A[i,(i+1)%n]=1
        A[(i+1)%n,i]=1
    return(A)
```

```python
def AdjHanoi(n=3):
    '''
    Returns adjacency matrix for the Tower of Hanoi graph with n disks
    '''
    Sl,Es=MakeHanoiGraph(n)
    m=len(Sl)
    A=np.zeros((m,m))
    for (i,j) in Es:
        A[i,j]=1
    return A
```

```python
def AdjBinary(n=6): # n must be even for square maze
    '''
    This constructs adjacency matrix for a binary maze with n branches 
    and 2^(n+1)-1 nodes according to our standard design.
    Operates at the resolution of nodes.
    Returns the adjacency matrix A(i,j)=1 if nodes i and j are neighbors.
    '''
    ma=NewMaze(n)
    nr=len(ma.ru) # number of runs
    A=np.zeros((nr,nr))
    for i in range(len(ma.ru)):  # for every run
        for c in ma.ch[i]:
            if c!=-1:
                A[i,c]=1
                A[c,i]=1
    return A
```

## Exact graph distance

```python
def Floyd(A): # Floyd–Warshall algorithm
    '''
    Computes shortest distance for all nodes starting from the adjacency matrix A
    Directed graphs: if 
        A[i,j] = connection from j to i
    then
        F[i,j] = shortest path from j to i
    '''
    A=np.array(A)
    n=len(A)
    F=(1-np.eye(n))*n-A*(n-1) # weight matrix: 0 on diagonal, 1 for connected nodes, n else
    for k in range(n):
        F=np.minimum(F,np.repeat(F[:,k][...,None],n,1)+np.repeat(F[k,:][None,...],n,0))
    return F
```

## Random walks

```python
def SystematicWalkBinary(l=6):
    '''
    Returns a systematic walk through all nodes and edges of the binary maze from 0 to 0
    '''
    A=AdjBinary(n=l)
    n=len(A)
    S=[np.nonzero(a==1)[0] for a in A.T] # S[i] = list of nodes you can step to from node i
    T=[np.zeros(len(S[i])) for i in range(n)] # marks nodes visited from i
    m=2*n-1 # length of optimal path through the maze
    P=[0]
    for s in range(1,m):
        i=P[s-1] # current node
        r=np.argwhere(T[i]==0) # indices of T[i] where value is 0, i.e. still unvisited
        j=np.argmax(S[i][r]) # index of largest value of S[i] yet unvisited
        T[i][r[j]]=1 # mark that node as visited
        P+=[S[i][r[j]][0]] # choose that node for the current step
    return P
```

```python
def SystematicWalkRing(n=50):
    '''
    Returns a systematic walk through all nodes and edges of the ring from 0 to 0
    '''
    return list(range(n))+[0]+list(range(n-1,-1,-1))
```

```python
def RandomWalkComplete(A,start=0,seed=0):
    '''
    Returns a random walk that ends once every edge has been covered
    A=adjacency matrix
    start=starting node
    seed=random seed
    '''
    np.random.seed(seed)
    n=len(A)
    S=[np.nonzero(a==1)[0] for a in A.T] # S[i] = list of nodes you can step to from node i
    T=A.copy()
    P=[start]
    m=np.sum(T>0) # number of edges there are
    e=0 # number of edges covered
    while e<m:
        i=P[-1] # current node
        j=np.random.choice(S[i])
        if T[j,i]==1:
            T[j,i]=0
            e+=1
        P+=[j]
    return P
```

```python
def DirectWalk(A,start,goal):
    '''
    Returns direct route from start to goal
    '''
    D=Floyd(A)
    S=[np.nonzero(a==1)[0] for a in A.T] # S[i] = list of nodes you can step to from node i
    p=start
    P=[p]
    while p!=goal:
        p=S[p][np.argmin(D[goal][S[p]])]
        P+=[p]
    return P
```

```python
def RandomWalk(A,start=0,seed=0,length=1000):
    '''
    Returns a random walk on the specified graph
    A=adjacency matrix
    start=starting node
    seed=random seed
    length=length
    '''
    np.random.seed(seed)
    S=[np.nonzero(a==1)[0] for a in A.T] # S[i] = list of nodes you can step to from node i
    P=[start]
    for _ in range(1,length):
        P+=[np.random.choice(S[P[-1]])]
    return P
```

```python
def ShortWalk(A,start=0,seed=0):
    '''
    A=adjacency
    s=random seed
    An attempt to create a short walk that covers all the edges of a graph
    At each node chooses randomly among the neighbor nodes that are least visited
    '''
    np.random.seed(seed)
    S=[np.nonzero(a==1)[0] for a in A.T] # S[i] = list of nodes you can step to from node i
    T=[np.zeros(len(s)) for s in S] # T[i] marks how often a neighbor has been visited from i
    p=start
    P=[p]
    e=np.sum(A) # number of edges that remain to be covered
    while e>0:        
        if np.min(T[p])==0: # there are neighbors that have not been visited
            j=np.random.choice(np.argwhere(T[p]==0)[0]) # pick one of those at random
            e-=1 # found another edge
        else:
            j=np.random.choice(len(T[p])) # otherwise pick a neighbor at random
        T[p][j]+=1 # mark that node as visited
        p=S[p][j] # choose that node for the current step
        P+=[p]
    return P
```

```python
def TestWalk(A,P):
    '''
    Check if walk P covers all the edges in A
    '''
    n=len(A)
    B=np.zeros((n,n))
    for i in range(len(P)-1):
        B[P[i],P[i+1]]=1
    return (B==A).all()
```

### Average occupancy of nodes during a long random walk
Define occupancy $o_i$ = probability that agent is at node $i$.

Because of the Markov character of the random walk, the occupancy satisfies

$$o_i = \sum_{j \ne i} o_j T_{ij} $$

where $T_{ij} $ is the transition matrix for the random walk. So $\bf{o}$ is the null vector of the matrix $\bf{T'}$, where

\begin{equation}
T'_{ij}=
\begin{cases}
  -1, & \text{if}\ i=j \\
  T_{ij}, & \text{otherwise}
\end{cases}
\end{equation}

`OccupRandomWalk(A)` computes that based on the adjacency matrix of the graph

```python
def OccupRandomWalk(A):
    '''
    returns the steady state occupancy of all nodes for a random walk on
    graph with adjacency matrix A
    A[i,j] = connection from j to i
    '''
    T=A/np.sum(A,axis=0) # Markov transition matrix
    np.fill_diagonal(T,-1)
    w,v=np.linalg.eig(T) # eigenvectors are in the columns of v!
    p=v[:,np.argmax(w.real)].real # eigenvector for the largest eigenvalue
    return p/np.sum(p) # normalize to sum=1
```

## TransMat

```python
def TransMat(A):
    '''
    returns the transition matrix, based on 
    A = adjacency matrix
    and equal transition probability for all edges from a given node
    '''
    return A/np.sum(A,axis=0)
```

## crit_gain

```python
def crit_gain(A):
    '''
    return critical gain value
    '''
    w,_=np.linalg.eigh(A) # eigenvectors are in the columns of v!
    return 1/np.max(w)
```

## map_lin

```python
def map_lin(M,ga):
    '''
    Return V = map output for every point cell activation
    V[:,j] = map output with agent at j
    '''
    return np.linalg.inv(1/ga*np.eye(len(M))-M)
```

## map_nonlin

```python
def map_nonlin(M,ga,mu=1):
    '''
    Return V = map output for every point cell activation
    V[:,j] = map output with agent at j
    '''
    def func(v):               # function whose zeroes we will find          
        return v-act(u+M@v)

    def act(x):                # activation function
        return np.minimum(ga*x,np.full(x.shape,ga*mu))

    n=len(M)                   # number of nodes
    v0=np.full(n,0.1)          # starting value for solving network equation
    U=np.eye(n)                # point cell activities
    V=[]
    for j,u in enumerate(U):   # for agent at node j
        V+=[fsolve(func,v0)]   # solve for the map output
    return np.array(V).T       # Transpose so that V[i,j] = map output i with agent at node j    
```

## P_max

```python
def P_max(g,no):
    '''
    Probability that a node will be chosen at an intersection, given readout noise no 
    Works for up to 3 node choices
    g[i] = goal signal at node i
    no = width of the uniform noise distribution
    Returns the probability p[i] that node i will have the largest readout
    '''
    if len(g)==1:
        return np.array([1])
    elif len(g)==2:
        s=np.argsort(g) # indices to sort g
        d=g[s[1]]-g[s[0]] # difference btw higher and lower value
        if (d>no): # no overlap between lower and higher, so higher wins
            p=np.array([0,1])
        else:
            p=np.array([(1-(d/no))/2,(1+(d/no))/2]) # take overlap into account
        return p[np.argsort(s)]        
    elif len(g)==3:
        s=np.argsort(g) # indices to sort g
        x,y,z=np.array(g)[s] # the 3 values in sorted order
        if z-x>no: # no overlap between the lowest and highest
            p=np.concatenate(([0],P_max((y,z),no))) # decide among the top 2
        else: # have overlap between the lowest and highest
            a=(x+no-z)/no
            b=(y-x)/no
            c=(z-y)/no
            p=np.array([a/3,a/3+b/2,a/3+b/2+c]) # take overlaps into account
        return p[np.argsort(s)]
```

## P_max_norm

```python
# probability for highest among 1 to 3 gaussian samples
def P_max_norm(g,no):
    '''
    Probability that a node will be chosen at an intersection, given gaussian readout noise 
    Works for up to 3 node choices
    g[i] = goal signal at node i
    no = 2*sigma of the gaussian noise distribution
    Returns the probability p[i] that node i will have the largest readout
    '''
    if len(g)==1:
        return np.array([1])
    elif len(g)==2:
        p0=norm.cdf(x=0,loc=g[1]-g[0],scale=np.sqrt(2)*(no/2)) # prob for first being higher
        return np.array([p0,1-p0])
    elif len(g)==3:
        rv=mvn(mean=[0,0],cov=[[1.0, 0.5],[0.5, 1.0]]) # to save time, could define this outside the function call
        k=1/np.sqrt(2)/(no/2) # scaling factor
        p0=rv.cdf([(g[0]-g[1])*k,(g[0]-g[2])*k])
        p1=rv.cdf([(g[1]-g[0])*k,(g[1]-g[2])*k])
        return np.array([p0,p1,1-p0-p1])
```

## Steps

```python
def Steps(A,F,E,no): 
    '''
    Returns avg number of steps from any given node to goals that may each include multiple nodes
    A = adjacency matrix
    E = goal signal E[i,j] = goal signal from node j to goal i
    F = feature strengths, F[i][j] = feature strength for goal i at node j
    no = noise added to the goal signal before readout
    '''
    p_lo=1e-6 # minimal transition probability. Avoids failed navigations.
    n=len(A) # number of nodes
    m=len(E) # number of goals
    S=[np.nonzero(a==1)[0] for a in A.T] # S[i] = list of nodes you can step to from node i
    T=np.zeros((m,n)) # T[i,j] = mean time from node j to goal i
    for k in range(m): # for every goal 
        e=E[k] # goal signal
        f=np.where(F[k]>0)[0] # feature locations for this goal
        Q=np.zeros((n,n)) # transition matrix
        for j in range(n):
            if j not in f: # if not yet at the goal
                x=e[S[j]] # goal signals at neighbor nodes
                p=(P_max_norm(x,no)+p_lo)/(1+len(x)*p_lo) # prob that each node will have highest readout
                for s,i in enumerate(S[j]):
                    Q[i,j]=p[s] # transition probabilities
        T[k]=T_avg(Q,f) # average time to that goal
        if np.min(T[k])<0: # failure in the average time calculation
            T[k]=1e6 # set this to something big
    return T
```

## Nav_Sim

```python
def Nav_Sim(A,F,E,no):
    '''
    Navigation based on goal signals. Average navigated distance.
    A = adjacency matrix
    F = goal locations (F[i] = list of target nodes for goal i)
    E = goal signal (E[i] = signal from every source node j to goal i)
    no = noise
    Returns:
    Ds = graph distance from every node to the goal
    Ld = avg navigated distance from every node to the goal
    Lm = mean of Ld for each graph distance value
    Ls = std of Ld for each graph distance value
    '''
    # avg length of navigated path
    L=Steps(A,F,E,no) # 

    # for each goal i and each node j find the minimum distance to the nodes in that goal's target list F[i]
    D=Floyd(A)
    Dg=np.array([np.min(D[f>0],axis=0) for f in F]) # locations where feature > 0 for each goal
    
    # Mean and std dev of distances
    Ds=sorted(list(set(Dg.flatten()))) # list of distinct distance values, sorted
    Ld=[L[Dg==d] for d in Ds] # list of navigated distances for each real distance
    
    return Ds,Ld
```

## Learn

```python
def Learn(A,F,Pars=(0.32,0.27,0.1,0),P=None,bi=False,M=None,G=None):
    '''
    Learn map and goal synapses during a random walk
    A = adjacency
    F = features, F[i][j] = feature strength for goal i at node j
    Pars = tuple of parameters, see below
    P = Trajectory of the walk
    bi = strengthen synapses bidirectionally
    M,G = map and goal synapses so far. If None these will start at 0.
    '''
    ga=Pars[0] # gain factor
    th=Pars[1] # threshold on activity
    al=Pars[2] # rate of goal learning
    de=Pars[3] # map and goal decay constant
    n=len(A) # number of nodes
    m=len(F) # number of goals
    notj=[[i for i in range(n) if i != j] for j in range(n)] # list of all i that are not j
    dmm=np.exp(-de) # multiplier for map synapse decay

    def learn_map():
        for j in np.where(u>th)[0]: # all cells with pre-before > th
            for i in notj[j]:
                if v[i]>th: # if post-after > th
                    M[i,j]=1 # potentiate that map synapse
                    if bi:
                        M[j,i]=1 # potentiate that map synapse
                else:
                    M[i,j]*=dmm # let that map synapse decay
                    if bi:
                        M[j,i]*=dmm # let that map synapse decay

    def learn_goal():
        for k in range(m): # for every goal
            e=G[k].dot(v) # goal signal
            f=F[k][p] # feature signal
            d=f-e # difference between feature signal and its prediction from the goal synapses
            if d>0: # if feature exceeds prediction
                G[k]+=al*v*d # increment those synapses by something prop to d and v # old line
#                 G[k]+=al*d*v/v.dot(v) # increment those synapses with normalization
            else:
                G[k]*=np.exp(-de*v) # let that synapse decay prop to v

    if M is None:
        M=np.zeros((n,n)) # erase map synapses
    if G is None:
        G=np.zeros((m,n)) # erase goal synapses
    if P is None:
        P=RandomWalkComplete(A,start=0,seed=0) # a random walk that covers all edges of the graph
    p=P[0] # start point of random walk
    v=np.linalg.inv(1/ga*np.eye(n)-M)[:,p] # compute map output
    learn_goal() # start this at very first node
    for p in P[1:]: # for each step along the walk
        u=v # previous map output
        v=np.linalg.inv(1/ga*np.eye(n)-M)[:,p] # compute new map output
        learn_map()
        learn_goal()
    return M,G
```

## Learn_nonlin

```python
def Learn_nonlin(A,F,Pars=(0.32,0.27,0.1,0),P=None,bi=False,M=None,G=None,mu=1):
    '''
    Learn map and goal synapses during a random walk
    A = adjacency
    F = features, F[i][j] = feature strength for goal i at node j
    Pars = tuple of parameters, see below
    P = Trajectory of the walk
    bi = strengthen synapses bidirectionally
    M,G = map and goal synapses so far. If None these will start at 0.
    '''
    from scipy.optimize import fsolve

    def func(v):               # function whose zeroes we will find          
        return v-act(u+M@v)

    def act(x):
        return np.minimum(ga*x,np.full(x.shape,ga*mu))
    
    n=len(A) # number of nodes
    m=len(F) # number of goals
    ga=Pars[0] # gain factor
    th=Pars[1] # threshold on activity
    al=Pars[2] # rate of goal learning
    de=Pars[3] # map and goal decay constant

    U=np.eye(n)   # point cell output
    v0=np.full(n,0.1)            # starting value for solving network equation
    notj=[[i for i in range(n) if i != j] for j in range(n)] # list of all i that are not j
    dmm=np.exp(-de) # multiplier for map synapse decay

    def learn_map():
        for j in np.where(s>th)[0]: # all cells with pre-before > th
            for i in notj[j]:
                if v[i]>th: # if post-after > th
                    M[i,j]=1 # potentiate that map synapse
                    if bi:
                        M[j,i]=1 # potentiate that map synapse
                else:
                    M[i,j]*=dmm # let that map synapse decay
                    if bi:
                        M[j,i]*=dmm # let that map synapse decay

    def learn_goal():
        for k in range(m): # for every goal
            e=G[k].dot(v) # goal signal
            f=F[k][p] # feature signal
            d=f-e # difference between feature signal and its prediction from the goal synapses
            if d>0: # if feature exceeds prediction
                G[k]+=al*v*d # increment those synapses by something prop to d and v # old line
#                 G[k]+=al*d*v/v.dot(v) # increment those synapses with normalization
            else:
                G[k]*=np.exp(-de*v) # let that synapse decay prop to v

    if M is None:
        M=np.zeros((n,n)) # erase map synapses
    if G is None:
        G=np.zeros((m,n)) # erase goal synapses
    if P is None:
        P=RandomWalkComplete(A,start=0,seed=0) # a random walk that covers all edges of the graph
    p=P[0] # start point of random walk

    u=U[p] # point cell output from node p
    v=fsolve(func,v0)
    
    learn_goal() # start this at very first node
    for p in P[1:]: # for each step along the walk
        s=v # previous map output
        u=U[p] # point cell output from node p
        v=fsolve(func,v0)
        learn_map()
        learn_goal()
    return M,G
```

## Perfect

```python
def Perfect(A,F,ga):
    '''
    Returns the optimal map and goal synapses and goal signals
    A = adjacency
    F = features, F[i][j] = feature strength for goal i at node j
    ga = gain
    '''
    M=A
    V=map_lin(M,ga) # V[i][j] = map output i with agent at node j
    G=F@(V.T/np.sum(V**2,axis=0).reshape(len(V),-1)) # G[i,j] = perfect goal synapses
    E=G@V
    E=np.minimum(E,1.0) # saturate all goal signals at 1; for a single target this should have no effect        
    return M,G,E
```

## Perfect_s

```python
def Perfect_s(A,F,ga):
    '''
    Like Perfect() but for a single goal synapse at the feature location
    Returns the optimal map and goal synapses and goal signals
    A = adjacency
    F = features, F[i][j] = feature strength for goal i at node j
    ga = gain
    '''
    M=A
    V=map_lin(M,ga) # V[i][j] = map output i with agent at node j
    G=F/np.diag(V) # synapses normalized so goal signal is equal to F at feature locations
    E=G@V # sum of map outputs at feature locations, weighted by feature strength
#     E=np.minimum(E,1.0) # saturate all goal signals at 1; for a single target this should have no effect        
    return M,G,E
```

## Perfect_nonlin

```python
def Perfect_nonlin(A,F,ga,mu=1):
    '''
    Returns the optimal map and goal synapses
    A = adjacency
    F = features, F[i][j] = feature strength for goal i at node j
    ga = gain
    mu = knee of saturating nonlinearity
    '''
    M=A
    V=map_nonlin(M,ga,mu) # V[i][j] = map output i with agent at node j
    G=F@(V.T/np.sum(V**2,axis=0).reshape(len(V),-1)) # G[i,j] = perfect goal synapses
    E=G@V
    E=np.minimum(E,1.0) # saturate all goal signals at 1; for a single target this should have no effect
    return M,G,E
```

## Goal

```python
# Goal signal with perfect map. 
def Goal(A,D,ga):
    '''
    Computes goal signal between any two points on a graph, given perfect map.
    Assumes that map synapses have acquired the adjacency matrix.
    Assumes goal signal is the scalar product of the two map outputs.
    Returns some statistics of those goal signals useful for plotting, see comments.
    A = adjacency matrix
    D = graph distance matrix
    ga = gain \gamma
    '''
    F=np.eye(len(A))
    _,_,E=Perfect(A,F,ga)
    Ds=list(set(D.flatten())) # list of distinct distance values
    Ed=[E[D==d] for d in Ds] # list of goal signal for each distance value
    Em=[np.mean(x) for x in Ed] # mean of goal signal for each distance value
    Es=[np.std(x) for x in Ed] # SD of goal signal for each distance value
    return Ds,Ed,Em,Es
```

```python
# Goal signal with perfect map. Single goal synapse.
def Goal_s(A,D,ga):
    '''
    Computes goal signal between any two points on a graph, given perfect map.
    Assumes that map synapses have acquired the adjacency matrix.
    Assumes goal signal is the output of the map cell at the goal location.
    Returns some statistics of those goal signals useful for plotting, see comments.
    A = adjacency matrix
    D = graph distance matrix
    ga = gain \gamma
    '''
    F=np.eye(len(A))
    _,_,E=Perfect_s(A,F,ga)
    Ds=list(set(D.flatten())) # list of distinct distance values
    Ed=[E[D==d] for d in Ds] # list of goal signal for each distance value
    Em=[np.mean(x) for x in Ed] # mean of goal signal for each distance value
    Es=[np.std(x) for x in Ed] # SD of goal signal for each distance value
    return Ds,Ed,Em,Es
```

## Goal_Range

```python
def Goal_Range(A,F,M,G,ga,no,fr=0.5):
    '''
    Measure the range of navigation on graph A using map M and goal synapses G towards goals F
    Do a linear fit to (fraction of routes off the identity) vs distance
    Interpolate the distance where that fraction is fr
    ga = gain
    no = noise
    '''
    V=np.linalg.inv(1/ga*np.eye(len(M))-M) # V[j]=map output with agent at j
    E=G@V # E[i,j] = goal signal from node j to goal i
    E=np.minimum(E,1.0) # saturate all goal signals at 1    
    Ds,Ls=Nav_Sim(A,F,E,no) # note Ds are sorted
    Y=np.array([len(np.where(ls>ds+0.5)[0])/len(ls) for ds,ls in zip(Ds,Ls)]) # fraction >d+0.5
    z=np.where(np.logical_and(Y[:-1]<fr,Y[1:]>=fr))[0] # find two points that straddle fr
    if len(z)==0: # if that never happens
        if Y[0]<fr: # fraction never exceeds fr, so return largest distance
            return Ds[-1] 
        else:  # fraction never below fr, so return 0
            return 0
    else:
        y0=Y[z[0]];y1=Y[z[0]+1]
        return Ds[z[0]]+(fr-y0)/(y1-y0)
```

## Goal_Range_nonlin

By introducing Gaussian noise and a minimum transition probability in Steps() none of the navigated distances are exactly the graph distance.

Introduce a new criterion for the range: Want the median navigated distance to be less than 0.5 steps longer than the graph distance.

```python
def Goal_Range_nonlin(A,F,M,G,ga,no,fr=0.5,mu=1):
    '''
    Same as Goal_Range, but with nonlinear activation curve
    Measure the range of navigation on graph A using map M and goal synapses G towards goals F
    Do a linear fit to (fraction of routes > d+0.5) vs distance
    Interpolate the distance where that fraction is fr
    ga = gain
    no = noise
    '''
    V=map_nonlin(M,ga,mu) # nonlinear map output
    E=G@V # goal signals
    E=np.minimum(E,1.0) # saturate all goal signals at 1    
    Ds,Ls=Nav_Sim(A,F,E,no) # note Ds are sorted
    Y=np.array([len(np.where(ls>ds+0.5)[0])/len(ls) for ds,ls in zip(Ds,Ls)]) # fraction >d+0.5
    z=np.where(np.logical_and(Y[:-1]<fr,Y[1:]>=fr))[0] # find two points that straddle fr
    if len(z)==0: # if that never happens
        if Y[0]<fr: # fraction never exceeds fr, so return largest distance
            return Ds[-1] 
        else:  # fraction never below fr, so return 0
            return 0
    else:
        y0=Y[z[0]];y1=Y[z[0]+1]
        return Ds[z[0]]+(fr-y0)/(y1-y0)
```

## T_avg

```python
def T_avg(Q,g):
    '''
    Q = graph transition matrix; Q[i,j] = probability of stepping to i when at j
    g = single goal node or list of goal nodes, works either way
    returns T[s] = average time from start node s to any of the goal nodes in g
    checks for failure when trying to invert the matrix 1-R
    '''
    n=len(Q)
    g=np.array([g]).flatten() # make an array, even if g is a single scalar
    m=len(g) # number of goal nodes
    R=np.delete(np.delete(Q,g,axis=0),g,axis=1) # remove rows and columns g from the transition matrix
    try:
        T=np.linalg.inv(np.identity(n-m)-R.T).dot(np.ones(n-m)) # solve
    except np.linalg.LinAlgError:
        T=-np.ones(n-m) # report failure with -1, this can happen if the graph is disconnected
    gi=g-np.arange(m) # indices for the insertion
    T=np.insert(T,gi,0) # add the missing rows back to the solution
    return T 
```

## do_nav

```python
def do_nav(A,ga=0.29,th=0.27,al=0.3,de=0,le=10000,bi=False):
    '''
    Compiles data on navigated distances after learning
    A = adjacency matrix
    ga, th, al, de = model parameters
    le = length of random walk for learning
    bi = bidirectional synapse learning?
    Returns data for the linear and log axis plots in the figure:
    E = goal signal for all pairs of nodes
    No = list of noise values used
    Ds = list of graph distances
    y50 = median of navigated distance for each graph distance
    C = error bars for 10/90 quantiles around the median
    '''
    # learn during exploration
    P=RandomWalk(A,0,0,le)
    n=len(A)
    F=np.eye(n) # place a goal at every point
    M,G=Learn(A,F,(ga,th,al,de),P,bi) # learn map and goal synapses from the random walk
    E=G@map_lin(M,ga) # All goal signals
    E=np.minimum(E,1.0) # saturate all goal signals at 1    
    
    # navigated length
    No=[0.005,0.01,0.02,0.05,0.1]
    data=[list(Nav_Sim(A,F,E,no)) for no in No]
    data+=[list(Nav_Sim(A,F,E,no=100))] # random (high noise)
    _,_,Ep=Perfect(A,F,ga)    
    data+=[list(Nav_Sim(A,F,Ep,no=0.01))] # perfect M and G
    
    # compute median ± 10/90 percentile
    Ds=data[0][0] # distances
    y50=np.array([[np.sort(l)[len(l)*50//100] for l in d[1]] for d in data])
    y10=np.array([[np.sort(l)[len(l)*10//100] for l in d[1]] for d in data])
    y90=np.array([[np.sort(l)[len(l)*90//100] for l in d[1]] for d in data])
    C=np.stack((y50-y10,y90-y50),axis=1) # error bars for 10/90 quantiles

    m=len(No)
    lin_data=(No,Ds,y50[:m],C[:m]) # data for the linear axis plot
    i=[m+1,1,m] # indices for perfect M,G with no=0.01; learned M,G with no=0.01; random
    log_data=(Ds[1:],y50[i,1:],C[i,:,1:]) # data for the log axis plot, omit the zero distance
    return E,lin_data,log_data
```

## do_range_nonlin

```python
# Compute goal signal range for all combinations of gain and threshold
def do_range_nonlin(A,GT,P,al=1,mu=1,bi=False):
    '''
    A = adjacency matrix
    GT = list of (gain,threshold) tuples
    P = walk for learning
    al = learning rate
    mu = saturation
    bi = bidirectional map learning?
    '''
    no=0.01 # Noise level
    de=0 # forgetting rate
    F=np.eye(len(A)) # a goal for every node
    Ra=[]
    for (ga,th) in GT:
        M,G=Learn_nonlin(A,F,(ga,th,al,de),P,mu=mu,bi=bi) # Learn M and G from random walk P
        Ra+=[Goal_Range_nonlin(A,F,M,G,ga,no,mu=mu)] # find range of navigation
    return np.array(Ra)
```

# Theory
Here we illustrate various analytical aspects of the goal signal. For this purpose we assume that the map and goal synapses have been acquired perfectly. No simulations with random numbers in this section.

## Greedy choice during navigation 

Say at an intersection there are 3 neighboring nodes available, with goal signals $g_i$. Those goal signals are corrupted by readout noise, distributed normally with standard deviation $\sigma=\epsilon/2$. What is the probability that each of the 3 nodes will be chosen? 

The problem is solved e.g. [here](https://mathoverflow.net/questions/207649/what-is-the-probability-that-a-gaussian-random-variable-is-the-largest-of-three).

Here the 3 Gaussian variables $Y_{\rm{A}},Y_{\rm{B}},Y_{\rm{C}}$ all have st dev = $\sigma$ and means $a, b, c$. The above prob is that $Y_{\rm{A}}$ is the largest. 

The function $L(h,k,\rho)$ is an integral of the bivariate gaussian distribution, with mean

$${\rm{mean}} = \begin{bmatrix}
0 \\
0 \\
\end{bmatrix}$$ 

and covariance matrix

$${\rm{cov}} = \begin{bmatrix}
1 & \rho \\
\rho & 1 \\
\end{bmatrix}$$

Note the integral is from $x$ to $\infty$, whereas the cumulative is usually expressed as the integral from $-\infty$ to $x$. So we just need to flip the sign of the first two arguments of $L$.

If there are just 2 samples $Y_{\rm{A}},Y_{\rm{B}}$, then the difference is normally distributed

$$d = Y_{\rm{A}}-Y_{\rm{B}} \sim \mathcal{N}(a-b,2\sigma^2)$$

So the probability that $Y_{\rm{A}}$ wins is 

$$P_{\rm{A}}=\int_0^\infty f(x) {\rm{d}}x$$

where $f(x)$ is the cumulative distribution function of the difference $d$

$$f(x)=\frac{1}{2 \sqrt{\pi} \sigma} {\rm{exp}} \left[ \frac{(x-(a-b))^2}{4 \sigma^2} \right]$$

We can compute both cumulatives numerically using `scipy.stats.norm`.

These calculations are done in `P_max_norm()`. Earlier versions of my code used a uniform distribution of width $\epsilon$. That case is handled by `P_max()`.

## Average navigated distance

Given a matrix of transition probabilities $Q_{ij}$ between nodes on a graph and a set $G$ of goal nodes, what is the average number of steps required from any start node $i$ to capture at one of the goal nodes?

This is a well-known problem, known for example as "discrete markov chain mean hitting time".

If $T_i$ is the mean number of steps to capture from node $i$, then by considering what happens after the first step, one finds

$$
T_i = 
\begin{cases}
    0, i \in G \\
    1+\sum_{k \ne i} T_k Q_{ki}, \textrm{else}
\end{cases}
$$

This leads to a solution involving simple matrix inversion.

`T_avg()` computes this.

```python
# illustrate T_avg with multiple targets on a ring graph. 
A=AdjRing(20)
Q=A/np.sum(A,axis=0) # Markov transition matrix
g=[2,7,13]
T=T_avg(Q,g)
plot(T);
```

`Steps()` computes the transition matrix for navigation walks based on a goal signal and noise. Then uses `T_avg()` to compute the average navigation length.

`Nav_Sim()` computes average navigation length between nodes and goals. Sorts the results by actual graph distance. This allows a comparison of real and navigated distance.

## Use transition matrix instead of adjacency matrix?

### Reviewer comment
From Reviewer #1: "My main concern with this paper is that the analysis of the critical gain value (gamma_c) is incomplete, making the implications of these analyses unclear. There are several different reasonable ways in which the Endotaxis map cell representations might be normalized, which I suspect may lead to different results. Specifically, the recurrent connections between map cells may either be an adjacency matrix, or a normalized transition matrix. In the current submission, the recurrent connections are an un-normalized adjacency matrix. In a previous preprint version of the Endotaxis manuscript, the recurrent connections between the map cells were learned using Oja's rule, which results in a normalized state-transition matrix (see "Appendix 5: Endotaxis model and the successor representation" in "Neural learning rules for generating flexible predictions and computing the successor representation", your reference 17). The authors state "In summary, this sensitivity analysis shows that the optimal parameter set for endotaxis does depend on the environment". Is this statement, and the other conclusions of the sensitivity analysis, still true if the learned recurrent connections are a properly normalized state-transition matrix?"

### Max eigenvalue of the transition matrix is 1.0
This is a well-known property of a transition matrix (also called "stochastic matrix"). 
Note the matrix is not symmetric, so it can have complex eigenvalues. But the one with the largest real part is equal to 1.0. See e.g. this proof



The proof also tells you that $|\lambda|<1$ for any complex eigenvalue. 

So the Taylor series for the resolvent of the transition matrix ${\bf{T}}$,

$$({\bf{1}} - \gamma {\bf{T}})^{-1} = {\bf{1}} + \gamma {\bf{T}} + \gamma^2 {\bf{T}}^2 + \dots$$

will converge as long as $\gamma < 1$.

In principle that means one could use the same gain $\gamma \approx 1$ on all graphs?


### Transition matrix at low gain also gives a goal signal exponentially related to graph distance
One can prove this in the same way as for the adjacency matrix. See my paper on the APSP algorithm. To confirm, here we plot the full range of goal signals for all node pairs with the same graph distance. The theory says that these ranges should not overlap for different graph distances, if the gain is chosen small enough.

```python
# goal signal vs graph distance
#     Binary maze. Map = Transition matrix.
#     Compute goal signal vs graph distance for every pair of start and end nodes.
#     Plot median ± full range for pairs with same graph distance
#     Various gain values possible

def Goal_1(M,D,ga):
    '''
    Like Goal(), but returns min and max
    '''
    F=np.eye(len(M))
    _,_,E=Perfect(M,F,ga)
    Ds=np.array(list(set(D.flatten()))) # list of distinct distance values
    Ed=[E[D==d] for d in Ds] # list of goal signal for each distance value
    Emed=np.array([np.median(x) for x in Ed]) # median of goal signal for each distance value
    Emin=np.array([np.min(x) for x in Ed]) # min of goal signal for each distance value
    Emax=np.array([np.max(x) for x in Ed]) # max of goal signal for each distance value
    return Ds,Ed,Emed,Emin,Emax

fig,ax=plt.subplots(1,1,figsize=(5,5))
A=AdjBinary(6) # graph
# A=AdjRing(51) # graph
M=TransMat(A)
# M=A
D=Floyd(A) # distance
Ga=[0.1] # gain values
col='rygcb'
data=[list(Goal_1(M,D,ga)) for ga in Ga] # compute the goal signals

Ds=data[0][0]
plot(Ds,[d[2] for d in data],fmts=[c+'o-' for c in col],
     legend=[f'gain={Ga[0]:.2f}',]+[f'{ga:.2f}' for ga in Ga[1:]],
     xlabel='Graph distance',
     ylabel='Goal signal',
     markersize=4,
     linewidth=1,
    yscale='log',
    loc='lower left',
    axes=ax,
);
for d,c in zip(data,col):
    ax.errorbar(d[0],d[2],yerr=[d[2]-d[3],d[4]-d[2]],
        fmt='none',
        ecolor=c,
         linewidth=1,       
    );
```

    - By our definition of range, namely >50% of navigations by the shortest route, the range is 8; whereas using the adjacency matrix the range is 12 (the full graph).
- At gain values below the critical value, the exponential relationship to graph distance is not as reliable as for the adjacency matrix (panel B of the figures).
- In the limit of very low gain, it is still true analytically that the goal signal represents the graph distance unambiguously.

## Use a single goal synapse?

### Reviewer comment
From reviewer #3: I note here that it was not clear what purpose, other than increasing the effective range of activation, is served by having the goal input weights set based on the activation levels when the goal is obtained. As demonstrated in the homing behaviour, it is sufficient to just have a goal connected to a single location for the mechanism to work (i.e., the activation at that location increases if the animal takes a step closer to it); and as demonstrated by adding a new graph connection, goal activation is immediately altered in an appropriate way to exploit a new shortcut, without the goal weights corresponding to this graph change needing to be relearnt.

One could make a learning rule for goal synapses that places a threshold on the map cell output so that only the directly activated map cell gets to make a synapse. Similar as for map learning. But is that a good idea?

```python
def do_binary_gain_m_s(ax,lo=0.1,hi=0.9):
    '''
    like do_binary_gain_m() but for single goal synapse
    Binary maze.
    Navigation based on perfect map and goal signals.
    Compute average navigation distance for every pair of start and end nodes.
    Plot against graph distance: median ± lo/hi quantiles
    Various gain values
    '''
    from matplotlib.ticker import MultipleLocator
    A=AdjBinary(6) # binary maze graph 
    F=np.eye(len(A)) # a goal for every node
    M=A # set map to adjacencies
    no=0.01 # noise
    Ga=[0.26,0.28,0.30,0.32,0.34] # gain values
    data=[]
    for ga in Ga:
        V=map_lin(M,ga) # V[i][j] = map output i with agent at node j
        G=F/np.diag(V) # synapses normalized so goal signal is equal to F at feature locations
        E=G@V # sum of map outputs at feature locations, weighted by feature strength
        data+=[Nav_Sim(A,F,E,no)] # compute average navigation distances
    Ds=data[0][0] # graph distances
    y50=[[np.sort(l)[int(0.5*len(l))] for l in d[1]] for d in data]
    ylo=[[np.sort(l)[int(lo*len(l))] for l in d[1]] for d in data]
    yhi=[[np.sort(l)[int(hi*len(l))] for l in d[1]] for d in data]
    col='rygcbm'
    plot(Ds,[Ds,]+y50,fmts=['k:',]+[c+'o-' for c in col],
         legend=[f'ideal',f'gain $\gamma$={Ga[0]:.2f}',]+[f'{ga:.2f}' for ga in Ga[1:]],
         xlabel='Graph distance',
         ylabel='Navigated distance',
         markersize=4,
         linewidth=1,
         axes=ax,
         loc='upper left',
    );
    ax.set_title(f'noise $\epsilon$={no:.2f}')
#     ax.text(6,105,f'noise $\epsilon$={no:.2f}')
    ax.xaxis.set_major_locator(MultipleLocator(5))
    ax.xaxis.set_minor_locator(MultipleLocator(1))
    ax.yaxis.set_major_locator(MultipleLocator(50))
    ax.yaxis.set_minor_locator(MultipleLocator(10))

    X=[d[0] for d in data]
    E=np.stack((np.array(y50)-np.array(ylo),np.array(yhi)-np.array(y50)),axis=1) # error bars
    for x,y,e,c in zip(X,y50,E,col):
        ax.errorbar(x,y,e,
            fmt='none',
            ecolor=c,
            linewidth=1,       
    );
```

# Learning
Here we simulate learning of map and targets during an exploratory random walk. At various points during the walk the target landscape or the connectivity changes.

`Learn_Sim()` implements the basic learning algorithm, with step by step recording of the synapse strengths and online changes in the structure of the environment.

## Simulations

```python
def Learn_Sim(ga,th,al,de,rs,bi=False):
    '''
    Learning both map and targets, simulation
    ga = gain
    th = threshold
    al = goal learning rate
    de = synaptic decay (not used here)
    rs = random seed
    bi = bidirectional synapse update?
    '''

    def set_graph():
        '''
        make changes to graph
        '''
        nonlocal N # binds N to the variable just outside this function
        if s==200: 
            F[0][4]=1 # target 1 appears
        elif s==400:
            F[0][8]=1 # target 2 appears for same goal
        elif s==600:
            A[6,13]=A[13,6]=1
            N=[np.nonzero(a==1)[0] for a in A.T] # N[i] = list of nodes you can step to from node i

    def learn_map():
        '''
        update the map synapses
        '''
        for j in np.where(u>th)[0]: # all cells with pre-before > th
            for i in notj[j]:
                if v[i]>th: # if post-after > th
                    M[i,j]=1 # potentiate that map synapse
                    if bi: # bidirectional update
                        M[j,i]=1 # same for the reverse synapse
                else:
                    M[i,j]*=dmm # let that map synapse decay
                    if bi: # bidirectional update
                        M[j,i]*=dmm # same for the reverse synapse

    def learn_goal():
        '''
        update the goal synapses
        '''
        for k in range(m): # for every goal
            e=G[k].dot(v) # goal signal
            f=F[k][p] # feature signal
            d=f-e # difference between feature signal and its prediction from the goal synapses
            if d>0: # if feature exceeds prediction
                G[k]+=al*d*v/v.dot(v) # increment those synapses by something prop to d and v
            else:
                G[k]*=np.exp(-de*v) # let that synapse decay prop to v

    A=AdjRing(14) # ring with 14 nodes
    le=1000 # length of random walk

    n=len(A)
    F=np.zeros((1,n)) # Feature strength, single goal
    m=len(F) # number of goals
    N=[np.nonzero(a==1)[0] for a in A.T] # N[i] = list of nodes you can step to from node i
    notj=[[i for i in range(n) if i != j] for j in range(n)] # list of all i that are not j
    dmm=np.exp(-de) # multiplier for map synapse decay

    Ps=[] # save random walk
    Ns=[] # save adjacencies
    Ms=[] # save the map synapses
    Gs=[] # save goal synapses
    M=np.zeros((n,n)) # erase map synapses
    G=np.zeros((1,n)) # erase goal synapses, single goal

    np.random.seed(rs) # set seed
    p=int(0) # start point of random walk
    v=np.linalg.inv(1/ga*np.eye(n)-M)[:,p] # compute map output
    for s in range(le): # for each step along the walk
        set_graph() # change the graph if needed
        u=v # previous map output
        p=np.random.choice(N[p]) # choose next node in random walk
        v=np.linalg.inv(1/ga*np.eye(n)-M)[:,p] # compute new map output
        learn_map() # update map synapses
        learn_goal() # update goal synapses
        Ps+=[p] # save random walk
        Ns+=[N.copy()] # save adjacencies
        Ms+=[M.copy()] # save map synapses
        Gs+=[G.copy()] # save goal synapses

    return Ps,Ns,Ms,Gs    
```

# Nonlinear activation

Here we consider the possible benefits of a nonlinear activation function for the map neurons.

Map output = $\vec v$

Point cell output = $\vec u$

Recurrent synapses = ${\bf{M}}$

Total input to map cells = $\vec w = \vec u + {\bf{M}} \vec v$

Activation function: $v_i = f(w_i) = f \left( u_i + \sum_j M_{ij} v_j \right)$

Start with a simple truncated linear response, namely

\begin{equation}
f(x)=
\begin{cases}
  \gamma x, & \text{if}\ x<\mu \\
  \gamma \mu, & \text{otherwise}
\end{cases}
\end{equation}

This is linear with slope $\gamma$ up to $x = \mu$, then flat.

For large $\mu$ it's simply a linear function.

## Relevant routines
`V=map_nonlin(M,ga,mu=1)` computes the output of the map neurons given ${\bf{M}}, \gamma, \mu$.

`M,G,E=Perfect_nonlin(A,F,ga,mu=1)` returns the map synapses M, goal synapses G, and goal signals E

`do_range_nonlin(A,GT,P,al=1,mu=1,bi=False)` computes the navigation range after learning on a random walk, for multiple (gain,threshold) tuples GT.

`do_range_nonlin_perf(A,Ga,mu=1)` does the same but assuming perfectly learned map and goal synapses.

## Range vs gain for lin vs nonlin activation

- Here we measure the navigation range after learning from a random walk, as measured in the figure on Tuning. For each of the 3 graphs:
    - compute the navigation range vs gain linear
    - compute range vs gain nonlinear
- Use re=30, al=0.1, as in the tuning figure

## Effects on learning

Consider an agent that has explored part of the environment and then steps into a new node. 
- With linear activation, the map output in the learned part of the environment is higher than at the new node, because of the contribution from recurrence.
- With saturating activation this doesn't happen.
- Illustrate this on the ring graph. 

```python
# ring; first few steps of the agent from the middle node; watch the map output
p1=[25,26,27,28,27,26,25,26,27,28,29] # path 
A=AdjRing(50)
n=len(A)
M=np.zeros((n,n))
ga=0.49 # highest gain with nonlin
Vs=[]
Vs_nonlin=[]
for t,j in enumerate(p1):
    V=map_nonlin(M,ga,mu=1000)[:,j] # linear
    Vs+=[V]
    V=map_nonlin(M,ga,mu=1)[:,j] # nonlinear
    Vs_nonlin+=[V]
    if t!=0:
        i=p1[t-1]
        M[i,j]=M[j,i]=1
Vs=np.array(Vs)
Vs_nonlin=np.array(Vs_nonlin)
```

```python
# linear
Y=Vs[-5:] # plot the last 5 map outputs
nc=len(Y)
# cm = plt.get_cmap('viridis')
cm = plt.get_cmap('tab10')
# cm = plt.get_cmap('gist_rainbow')
fig,ax=plt.subplots(1,1,figsize=(4,2))
ax.set_prop_cycle(color=cm(np.linspace(0,1,nc)))
for i in range(nc):
    ax.plot(Y[i],marker='o',ms=3)
ax.plot([24,30],[Y[-1,29],Y[-1,29]],color='k',ls=':',lw=1)
ax.set_xlim(23,31);
ax.set_xticks(np.arange(23,32));
ax.set_yticks([0,0.5,1]);
ax.set_xticklabels(list(range(0,9)));
ax.set_xlabel('Node');
ax.set_ylabel('Map output');
```

```python
# nonlinear
Y=Vs_nonlin[-5:] # plot the last 5 map outputs
nc=len(Y)
cm = plt.get_cmap('tab10')
# cm = plt.get_cmap('gist_rainbow')
fig,ax=plt.subplots(1,1,figsize=(4,2))
ax.set_prop_cycle(color=cm(np.linspace(0,1,nc)))
for i in range(nc):
    ax.plot(Y[i],marker='o',ms=3)
ax.plot([24,30],[Y[-1,29],Y[-1,29]],color='k',ls=':',lw=1)
ax.set_xlim(23,31);
ax.set_xticks(np.arange(23,32));
ax.set_xticklabels(list(range(0,9)));
ax.set_xlabel('Node');
ax.set_ylabel('Map output');
```

In the linear case, the output of the map grows as more synapses get added. To properly identify which map cell gets directly activated by a point cell one must choose a threshold below the peak signal on the first step (dotted line). But a few steps later, that threshold would also select inappropriate map cells, and so learn the wrong adjacency matrix.

In the saturating case, the peak map output is always the same, so one can use a constant threshold to identify the map cell corresponding to the agent's location.

## Drawing routines 

```python
def draw_bin_goal_2(ax,E): 
    n=len(E)
    le=int(np.log(n+1)/np.log(2))-1
    I=np.zeros((2**le,le+1))
    for l in range(le+1):
        for k in range(2**l):
            I[k*2**(le-l):(k+1)*2**(le-l),l]=E[2**l+k-1]
    ax.imshow(I,aspect='auto',interpolation='none');
    for l in range(le+1):
        for k in range(2**l-1):
            y=(k+1)*2**(le-l)-0.5
            plot([l-0.5,l+0.5],[y,y],fmts=['w-'],linewidth=0.5,axes=ax);
    ax.set_yticks([]);
    ax.set_xticks([]);
```

```python
def draw_home_traj(ax):
    from scipy.signal import savgol_filter
    from matplotlib.collections import LineCollection

    def path(p0,p1,de):
        return np.linspace(p0,p1,
                           int(np.linalg.norm(np.array(p1)-np.array(p0))/de)
                           ,endpoint=True)

    # draw maze
    ma=NewMaze(6)
    ax.plot(ma.wa[:,0],ma.wa[:,1],'-',linewidth=1,color='gray')
    X=[];Y=[]
    X+=[[0,6,8,14]];Y+=[[1,5,9,13]]
    X+=[[3,11]];Y+=[[0,6,8,14]]
    X+=[[0,1,2,12,13,14]];Y+=[[3,11]]
    X+=[[7]];Y+=[[0,1,2,12,13,14]]
    X+=[[8,9,10,11,12,13,14]];Y+=[[7]]
    for x,y in zip(X,Y):
        for h in x:
            for v in y:
                rect=patches.Rectangle((h-0.5,v-0.5),1,1,edgecolor='None',facecolor='lightgray')
                ax.add_patch(rect);
    ax.set_aspect(1)

    # make trajectory
    s=0.18
    de=0.1
    # node sequence [0,1,4,9,20,41,84,41,20,9,4,10,22,46,94]
    X1=np.array([-0.5-s,7,7,11,11,13,13,14],dtype=float)-s
    Y1=np.array([7,7,11,11,13,13,14,14],dtype=float)+s
    X2=np.array([14,13,13,11,11,11,13,13,14],dtype=float)+s
    Y2=np.array([14,14,13,13,11,9,9,8,8],dtype=float)-s
    P=np.array([np.concatenate((X1,X2)),np.concatenate((Y1,Y2))]).T

    L=[P[0]]
    for i in range(len(P)-1):
        L=np.append(L,path(P[i],P[i+1],de)[1:],axis=0)
    # smooth trajectory
    Lf=savgol_filter(L,5,1,axis=0)

    # draw trajectory
    segs=np.stack((Lf[:-1],Lf[1:]),axis=1)
    c=np.arange(len(segs))
    lc = LineCollection(segs,array=c,linewidths=3)
    ax.add_collection(lc);
    
    # draw labels
    X=[6.8,10.1,14.6,14.6]
    Y=[7,10.9,14.5,8.5]
    L=[1,2,3,4]
    for x,y,l in zip(X,Y,L):
        ax.text(x,y,l,ha='left',va='top',fontsize=12)
```

# Patrolling
In patrolling the agent ascends on the signal of a patrol cell. The patrol cell gets synapses of equal strength from every map cell. Those synapses are normalized so the maximal range of the patrol signal is 1. As before, that signal gets corrupted by noise of width $\epsilon$ before readout. 

The other new ingredient is habituation of the point cells. On every visit to a point cell, its output declines by a factor of $e^{-\beta}$. And on every time step the habituation decays by a factor $e^{-1/\tau}$.


## Patrol algorithm

`Patrol()` implements the patrolling algorithm, including the learning of map and goal synapses during the patrol. 

```python
def Patrol(A,F,M,G,h,no,ga,th,al,be,ta,st,le,rs,learn=True):
    '''
    Implements patrolling, starting from the given state of the map M and the goal synapses G
    Continues to learn M and G, if learn==True
    h = habituation state of the point cells at the start
    no,ga,th,al,be,ta = model parameters
    st = starting node for the patrol
    le = length of the patrol
    rs = random seed
    Returns:
    Q = walk produced by the patrol
    M, G, h = new values of map synapses, goal synapses, and habituation
    '''
    def learn_map(): # note bidirectional potentiation
        for j in np.where(v1>th)[0]: # all cells with pre-before > th
            for i in notj[j]: # no self-synapses
                if v[i]>th: # if post-after > th
                    M[i,j]=M[j,i]=1 # potentiate map synapse in both directions

    def learn_goal():
        for k in range(m): # for every goal
            f=F[k][q] # feature signal at current node
            if f>0: # only modify goal synapses if feature signal is present
                e=G[k].dot(v) # goal signal
                d=f-e # difference between feature signal and its prediction from the goal synapses
                G[k]+=al*v*d # change those goal synapses by something prop to d and v - old

    def step_random():
        return np.random.choice(N[q])

    def step_patrol():
        E=np.sum(np.linalg.inv(1/ga*np.eye(n)-A),axis=0)*h/Emax # patrol signal from every location, max=1
        x=E[N[q]] # goal signals at neighbor nodes
        p=(P_max_norm(x,no)+p_lo)/(1+len(x)*p_lo) # prob that each node will have highest readout
        return np.random.choice(N[q],p=p) # choose the new node

    n=len(A)
    m=len(F)
    N=[np.nonzero(a==1)[0] for a in A.T] # N[i] = list of nodes you can step to from node i
    notj=[[i for i in range(n) if i != j] for j in range(n)] # list of all i that are not j
    Emax=np.max(np.sum(np.linalg.inv(1/ga*np.eye(n)-A),axis=0)) # largest possible sum over map output 
    # use this to normalize the patrol signal so the noise is scaled appropriately
    p_lo=1e-6 # min transition probability
    
    ha=np.exp(-be) # habituation factor on every visit to a point cell
    dh=np.exp(-1/ta) # decay of the habituation on every time step

    np.random.seed(rs)
    q=st # start node
    Q=[q]
    v=np.linalg.inv(1/ga*np.eye(n)-M)[:,q]*h[q] # map output from current location
    if learn==True:
        learn_goal() # update the goal synapses at very first node
    while len(Q)<le:
        v1=v # previous map output
        q=step_patrol() # ascending on the patrol signal
        Q+=[q] # record the new step
        v=np.linalg.inv(1/ga*np.eye(n)-M)[:,q]*h[q] # compute new map output
        if learn:
            learn_map() # update the map synapses
            learn_goal() # update the goal synapses
        h=1-(1-h)*dh # point cell recovery
        h[q]*=ha # point cell habituation    
    return Q,M,G,h
```

```python
# fit function
def func(x,a,b): # the 2-parameter fit function
    z=x/a
    return 64*(1-1/(1+(z+b*z**3)/(1+b)))
```

```python
# new node analysis
def new_nodes(Q,X=None):
    '''
    Q=sequence of nodes
    slides a window of various widths across the sequence
    X=list of window widths. If omitted makes its own list with 50 values.
    returns y=mean number of distinct nodes in that window
    '''
    n=len(Q)
    Y=[]
    if X is None:
        X=np.floor(np.exp(np.log(n)*np.arange(0,1,0.02))).astype(int)
    for x in X:
        if x<1:
            Y+=[0]
        else:
            R=[Q[i:i+x] for i in range(0,max(n-x,0)+1,max(x//2,1))]
            Y+=[np.mean([len(set(r)) for r in R])]
    return X,Y
```

# Forgetting

Here we simulate learning of map and goals with a learning rule that includes active forgetting. Map synapses decay if presynaptic activity is high but postsynaptic is low. Similarly goal synapses decay when the prediction from the map input is high but the feature input is low.

## Simulations

```python
# Learning both map and targets on ring graph, including forgetting

def set_graph():
    '''
    make changes to graph
    '''
    global A,N
    if s==0: 
        F[0][4]=1 # target appears
    elif s==200:
        A[4,11]=A[11,4]=1 # new link to target
        N=[np.nonzero(a==1)[0] for a in A.T] # N[i] = list of nodes you can step to from node i
    elif s==400:
        A[4,11]=A[11,4]=0 # link disappears
        N=[np.nonzero(a==1)[0] for a in A.T] # N[i] = list of nodes you can step to from node i

def learn_map():
    for j in np.where(u>th)[0]: # all cells with pre-before > th
        for i in notj[j]:
            if v[i]>th: # if post-after > th
                M[i,j]=1 # potentiate that map synapse
                M[j,i]=1 # and the reverse synapse; remove for directed graph
            else:
                M[i,j]*=dmm # let that map synapse decay
                M[j,i]*=dmm # let that map synapse decay

def learn_goal():
    for k in range(m): # for every goal
        e=G[k].dot(v) # goal signal
        f=F[k][p] # feature signal
        d=f-e # difference between feature signal and its prediction from the goal synapses
        if d>0: # if feature exceeds prediction
            G[k]+=al*v*d # increment that synapse by something prop to d and v - old
        else:
            G[k]*=np.exp(-de*v) # let that synapse decay prop to v

A=AdjRing(14)
ga=0.32 # gain factor
th=0.27 # threshold on map activity
al=0.3 # rate of goal learning
de=0.1 # synaptic decay factor \delta
le=1000 # length of random walk
rs=2 # random seed

n=len(A)
F=np.zeros((1,n)) # Feature strength, single goal
m=len(F) # number of goals
N=[np.nonzero(a==1)[0] for a in A.T] # N[i] = list of nodes you can step to from node i
notj=[[i for i in range(n) if i != j] for j in range(n)] # list of all i that are not j
dmm=np.exp(-de) # multiplier for map synapse decay

Ps=[] # save random walk
Ns=[] # save adjacencies
Ms=[] # save the map synapses
Gs=[] # save goal synapses
M=np.zeros((n,n)) # erase map synapses
G=np.zeros((1,n)) # erase goal synapses, single goal

np.random.seed(rs)
p=int(0) # start point of random walk
v=np.linalg.inv(1/ga*np.eye(n)-M)[:,p] # compute map output
for s in range(le): # for each step along the walk
    set_graph()
    u=v # previous map output
    p=np.random.choice(N[p]) # choose next node in random walk
    v=np.linalg.inv(1/ga*np.eye(n)-M)[:,p] # compute new map output
    learn_map()
    learn_goal()
    Ps+=[p] # save random walk
    Ns+=[N.copy()]
    Ms+=[M.copy()] # save map synapses
    Gs+=[G.copy()] # save goal synapses

filename='sims/forget'
with open(fn,'wb') as f:
    pickle.dump(((ga,th,al,de,rs),Ps,Ns,Ms,Gs), f)      
```

# Habituation

Here we analyze the effects that point cell habituation has on the performance of endotaxis. There are three possible concerns one might have:
- Map learning: The learning of map synapses happens when both pre- and post-synaptic cell fire above a certain threshold. With habituating point cells, these threshold crossings happen more rarely, only when both neurons are sufficiently "fresh". One can't really lower the threshold, because that leads to inappropriate synapses. In practice these effects just require somewhat longer exploratory walks for complete learning of the map.
- Goal learning: Habituation implies that the map output could be quite different on subsequent visits to the goal location, depending on the prior history of the walk. That means very high values of the learning rate $\alpha$ are less useful. In practice, $\alpha=0.3$ and $\alpha=1$ give good results.
- Navigation after learning. Again, one may worry that the goal signal now depends on history in a way that interferes with proper navigation. In practice the simulations show that it makes little difference.


We used values for habituation $\beta = 1.2$ and recovery time $\tau = 100 \, \rm{steps}$ that produce good patrolling behavior. 

`Learn_Hab()` implements learning of map and goal signals during an exploratory walk. In the reported simulation map synapses get strengthened bidirectionally.

`Nav_Hab()` implements navigation between any two nodes using the learned goal signal. Here the question comes up: What is the habituation state of the point cells when navigation starts? Two options come to mind:
1. Start the point cells at full gain, i.e. $h_i=1$.
2. Start them in the average habituation state expected after a long random walk. `Avg_Hab()` computes that.

In practice it makes little difference. Option (2) produces slightly worse navigation, so we conservatively reported that.

`do_Hab()` combines the two simulations for various parameters values and saves results to file for subsequent plotting.

## Average habituation state

At each visit to a point cell, its gain declines by a factor 

$$H = {\rm{e}}^{-\beta}$$

At each time step, the gain of all point cells recovers by a factor 

$$R = {\rm{e}}^{-1/\tau}$$

We want to know the average habituation state of each point cell after a long random walk. Say we knew the average occupancy of each node $o_j$. Then the average habituation state of the point cell $h_i$ at time $t+1$ satisfies

$$h_i(t+1) = o_i h_i(t) H + (1 - o_i) (1 - (1-h_i(t)) R) $$

In steady state, $h_i(t+1) = h_i(t)$, so

$$h_i = \frac{(1-o_i)(1-R)}{(1-o_i H)-(1-o_i) R}$$

`OccupRandomWalk()` computes the average occupancy

`Avg_Hab()` computes the average habituation state

```python
def Avg_Hab(A,ga,be,ta):
    '''
    Returns the average habituation state of point cells after a long random walk
    A = adjacency
    ga = gain \gamma
    be = habituation \beta
    ta = recovery time \tau
    Returns
    h = haverage habituation state of all point cells
    '''
    ha=np.exp(-be) # habituation factor on every visit to a point cell
    dh=np.exp(-1/ta) # decay of the habituation on every time step
    o=OccupRandomWalk(A) # average occupancy of every node
    h=(1-o)*(1-dh)/((1-o*ha)-(1-o)*dh) # average habituation of every node
    return h
```

## Learning map and goals with habituation

```python
# Learning routine including habituation
def Learn_Hab(A,F,ga,th,al,be,ta,P=None,bi=False):
    '''
    A = adjacency matrix
    F = features, F[i][j] = feature strength for goal i at node j
    ga,th,al,be,ta = model parameters
    P = walk for learning
    bi = whether to strengthen synapses bidirectionally
    '''
    def learn_map():
        for j in np.where(v1>th)[0]: # all cells with pre-before > th
            for i in notj[j]:
                if v[i]>th: # if post-after > th
                    M[i,j]=1 # potentiate that map synapse
                    if bi:
                        M[j,i]=1 # potentiate that map synapse

    def learn_goal():
        for k in range(m): # for every goal
            f=F[k][p] # feature signal
            if f>0: # only modify goal synapses if feature signal is present
                e=G[k].dot(v) # goal signal
                d=f-e # difference between feature signal and its prediction from the goal synapses
                G[k]=np.maximum(G[k]+al*v*d,0) # change those goal synapses by something prop to d and v; cap at zero

    n=len(A) # number of nodes
    m=len(F) # number of goals
    notj=[[i for i in range(n) if i != j] for j in range(n)] # list of all i that are not j

    ha=np.exp(-be) # habituation per visit
    dh=np.exp(-1/ta) # decay of habituation per time step

    M=np.zeros((n,n)) # erase map synapses
    G=np.zeros((m,n)) # erase goal synapses
    h=np.ones(n) # point cells start out with full gain
    p=P[0] # start point of random walk
    v=np.linalg.inv(1/ga*np.eye(n)-M)[:,p]*h[p] # map output from current location
    learn_goal() # start this at very first node
    for p in P[1:]: # for each step along the walk
        v1=v # previous map output
        v=np.linalg.inv(1/ga*np.eye(n)-M)[:,p]*h[p] # compute new map output
        learn_map()
        learn_goal()
        h=1-(1-h)*dh # recovery
        h[p]*=ha # habituation
    return M,G,h # learned map, goals, and last habituation state
```

## Navigating with habituation

```python
# Navigation between two points including habituation
def Nav_Hab(A,M,G,h,ga,be,ta,no,rs,x,y):
    '''
    Navigate from node x to y following a single goal signal G (1D array)
    Map and goal signal stay fixed during navigation
    A=adjacency
    M=map synapses
    G=goal synapses
    h=habituation state at start
    ga,be,ta,no=model parameters
    rs=random seed
    x=start node
    y=goal node
    Returns the walk including x and y
    '''
    def step_nav():
        E=W*h # goal signal from every location including habituation
        x=E[N[q]] # goal signals at neighbor nodes
        p=(P_max_norm(x,no)+p_lo)/(1+len(x)*p_lo) # prob that each node will have highest readout
        return np.random.choice(N[q],p=p) # choose the new node

    ha=np.exp(-be) # habituation per visit
    dh=np.exp(-1/ta) # decay of habituation per time step
    np.random.seed(rs)
    N=[np.nonzero(a==1)[0] for a in A.T] # N[i] = list of nodes you can step to from node i
    n=len(A)
    p_lo=1e-6 # min transition probability
    W=G@np.linalg.inv(1/ga*np.eye(n)-M) # goal signal from every node, not including habituation
    q=x # start walk
    Q=[q]
    while q!=y:
        q=step_nav() # ascend on the goal signal
        Q+=[q] # record the new step
        h=1-(1-h)*dh # point cell recovery
        h[q]*=ha # point cell habituation    
    return np.array(Q)
```

## Simulations

```python
# Simulate learning and navigation including point cell habituation at various noise levels
# Use the same routine for a control without habituation
def do_Hab(be,ta):
    '''
    Simulates learning and habituation on the binary maze
    be = habituation
    ta = recovery time 
    all other parameters external
    '''
    # learn the map and goal signals from a random walk
    rs=0
    P=RandomWalk(A,start=0,seed=rs,length=le)
    M,G,h=Learn_Hab(A,F,ga,th,al,be,ta,P,bi=True) 

    # then use the learned map and goal signals to navigate once for every pair of nodes
    # do this for different noise values
    Q=[]
    for no in No:
        Qn=[]
        for x in range(n): # all start nodes
            for y in range(n): # all goal nodes
                if y!= x:
                    rs1=rs+(x+1)*(y+1) # different random seed for each navigation
                    h=Avg_Hab(A,ga,be,ta) # set h to average after random walk
#                     h=np.ones(n) # start point cells at full gain; agent has rested
                    Qn+=[[x,y,len(Nav_Hab(A,M,G[y],h,ga,be,ta,no,rs1,x,y))-1]] # compute navigated distance
#             print('\t',x) # report progress 
#         print(f'no = {no}') # report progress
        Q+=[Qn]
    return np.array(Q)

A=AdjBinary(6)
No=[0.005,0.01,0.02,0.05,0.1] # same parameters as in the Navigation figure 
ga=0.33
th=0.30 
al=0.1
le=30000
n=len(A)
F=np.eye(len(A)) # all nodes are goals

be=1.2 # habituation
ta=100 # recovery time
Q=do_Hab(be,ta)
fn=f'sims/habituation'
with open(fn,'wb') as f:
    pickle.dump((No,Q),f)      
    
be=0 # no habituation
ta=100
Q=do_Hab(be,ta)
fn=f'sims/habituation_nh'
with open(fn,'wb') as f:
    pickle.dump((No,Q),f)      
```

## Figure

```python
def draw_nav_hab(ax,fn):
    '''
    Draw results of navigation based on simulations saved to file fn
    fn = file path from project directory
    le = whether to add legend
    '''
    # load data
    with open(fn,'rb') as f:
        (No,Q)=pickle.load(f)      
    # sort results by graph distance
    A=AdjBinary(6)
    D=Floyd(A)
    Ds=np.array([D[x,y] for [x,y] in Q[0,:,0:2]]).astype(int) # distances
    Dm=np.max(Ds).astype(int) # largest distance
    L=[[] for _ in range(Dm)]
    for i,d in enumerate(Ds):
        L[d-1]+=[Q[:,i,2]]
    for i,l in enumerate(L):
        L[i]=np.sort(np.array(l).T,axis=1) # L[distance][noise][pair]
    # find the quantiles
    q=np.array([0.1,0.5,0.9]) # desired quantiles
    Lq=np.array([[d[(q*len(d)).astype(int)] for d in e] for e in L]) # Lq[distance,noise,quantile]
    # plot the quantiles
    x=np.arange(1,Dm+1) # x-axis = graph distance
    ylo,y50,yhi=Lq[:,:,0].T,Lq[:,:,1].T,Lq[:,:,2].T # quantiles of navigated distance [noise,distance]
    col='rgbycm'
    plot(x,x,fmts=['k:',],axes=ax,);
    plot(x,y50,fmts=[c+'o-' for c in col],
         legend=[f'ideal',f'noise $\epsilon$={No[0]:.3f}',]+[f'{no:.3f}' for no in No[1:]],
         xlabel='Graph distance',
         ylabel='Navigated distance',
         markersize=4,
         linewidth=1,
         axes=ax,
         loc='upper left',
         ylim=[-5,50],
         xlim=[-0.5,12.5],
    );
    ax.xaxis.set_major_locator(MultipleLocator(5))
    ax.xaxis.set_minor_locator(MultipleLocator(1))
    ax.yaxis.set_major_locator(MultipleLocator(50))
    ax.yaxis.set_minor_locator(MultipleLocator(10))
    C=np.stack((np.array(y50)-np.array(ylo),np.array(yhi)-np.array(y50)),axis=1)
    for y,e,c in zip(y50,C,col):
        ax.errorbar(x,y,e,
            fmt='none',
            ecolor=c,
            linewidth=1,       
    );
```

```python
# draw the figure, random walk of length 30000
le=30000
fig,ax=plt.subplots(1,2,figsize=(7,3))
draw_nav_hab(ax[0],'sims/habituation')
draw_nav_hab(ax[1],'sims/habituation_nh')
ax[1].get_legend().remove()
ax[1].set_ylabel('')
ax[0].set_title(r'Habituation: $\beta=1.2, \tau=100$')
ax[1].set_title('No habituation')
plt.tight_layout()
plt.savefig('figs/Habituation.pdf')
```
